## 4. Data Cleaning

### 4.1 Missing Values

```python
missing = df.isnull().sum()
missing_pct = (missing / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing,
    'Missing %': missing_pct
})

missing_df = missing_df[missing_df['Missing Count'] > 0].sort_values(
    'Missing Count',
    ascending=False
)

print(missing_df)
fig, ax = plt.subplots(figsize=(8, 4))

missing_df['Missing %'].plot(
    kind='barh',
    ax=ax,
    color='steelblue',
    edgecolor='white'
)

ax.set_xlabel('Missing Percentage (%)')
ax.set_ylabel('Columns')
ax.set_title('Missing Values by Column')

for i, v in enumerate(missing_df['Missing %']):
    ax.text(v + 0.3, i, f'{v}%', va='center', fontsize=9)

plt.tight_layout()

# Save graph to images folder
plt.savefig("../images/missing_values_by_column.png", dpi=300, bbox_inches="tight")

plt.show()
```

### 4.2 Handle Missing Values

```python
# Drop columns with excessive missing data (>50%)
df.drop(columns=['dish_liked', 'phone', 'url', 'address', 'reviews_list', 'menu_item'], inplace=True)

# Drop remaining rows with nulls (location, rest_type, cuisines, approx_cost, rate)
df.dropna(inplace=True)

print(f'Shape after handling missing values: {df.shape}')
print(f'Remaining nulls: {df.isnull().sum().sum()}')
```

### 4.3 Remove Duplicates

```python
print(f'Duplicate rows before: {df.duplicated().sum()}')
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
print(f'Duplicate rows after : {df.duplicated().sum()}')
print(f'Shape after deduplication: {df.shape}')
```

### 4.4 Correct Data Types

```python
# Clean 'rate': '4.1/5' → 4.1
df['rate'] = df['rate'].astype(str).str.replace('/5', '', regex=False).str.strip()
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
df.dropna(subset=['rate'], inplace=True)

# Clean 'approx_cost(for two people)': remove commas → float
df['approx_cost(for two people)'] = (
    df['approx_cost(for two people)']
    .astype(str)
    .str.replace(',', '', regex=False)
    .pipe(pd.to_numeric, errors='coerce')
)
df.dropna(subset=['approx_cost(for two people)'], inplace=True)

# Rename for convenience
df.rename(columns={
    'approx_cost(for two people)': 'cost_for_two',
    'listed_in(type)': 'listed_type',
    'listed_in(city)': 'listed_city'
}, inplace=True)

print('Data types after cleaning:')
print(df.dtypes)
print(f'\nFinal shape: {df.shape}')
```

## 5. Outlier Detection

### 5.1 Boxplots — Visual Detection

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, col, color in zip(
    axes,
    ['rate', 'votes', 'cost_for_two'],
    ['#5c85d6', '#e07b54', '#6abf69']
):
    ax.boxplot(
        df[col].dropna(),
        patch_artist=True,
        boxprops=dict(facecolor=color, alpha=0.6),
        medianprops=dict(color='black', linewidth=2)
    )
    ax.set_title(col.capitalize(), fontsize=11)
    ax.set_xticks([])

plt.suptitle('Boxplots for Outlier Detection', fontsize=13, y=1.02)

plt.tight_layout()

# Save image
plt.savefig("../images/boxplots_outlier_detection.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

### 5.2 Handle Outliers — IQR Method

```python
def remove_outliers_iqr(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = len(data)
    data = data[(data[col] >= lower) & (data[col] <= upper)]
    print(f'{col}: removed {before - len(data):,} outliers  (lower={lower:.1f}, upper={upper:.1f})')
    return data

df = remove_outliers_iqr(df, 'votes')
df = remove_outliers_iqr(df, 'cost_for_two')
df.reset_index(drop=True, inplace=True)
print(f'\nShape after outlier removal: {df.shape}')
```

### 5.3 Boxplots — After Outlier Removal

```python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

for ax, col, color in zip(
    axes,
    ['rate', 'votes', 'cost_for_two'],
    ['#5c85d6', '#e07b54', '#6abf69']
):
    ax.boxplot(
        df[col].dropna(),
        patch_artist=True,
        boxprops=dict(facecolor=color, alpha=0.6),
        medianprops=dict(color='black', linewidth=2)
    )
    
    ax.set_title(col.replace('_', ' ').title(), fontsize=11)
    ax.set_xticks([])

plt.suptitle('Boxplots After Outlier Removal', fontsize=13, y=1.02)

plt.tight_layout()

# Save graph to images folder
plt.savefig(
    "../images/boxplots_after_outlier_removal.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

## 8. Correlation Analysis

### 8.1 Correlation Heatmap

```python
num_cols = df.select_dtypes(include='number')
corr_matrix = num_cols.corr()

fig, ax = plt.subplots(figsize=(7, 5))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))
sns.heatmap(corr_matrix, annot=True, fmt='.2f', cmap='coolwarm',
            mask=mask, vmin=-1, vmax=1,
            linewidths=0.5, square=True, ax=ax,
            annot_kws={'size': 11})
ax.set_title('Correlation Heatmap (Numeric Features)')

plt.tight_layout()

# Save graph to images folder
plt.savefig("../images/correlation_heatmap.png", dpi=300, bbox_inches="tight")

plt.show()
```

### 8.2 Identify Highly / Negatively Correlated Features

```python
corr_pairs = (
    corr_matrix.where(np.tril(np.ones(corr_matrix.shape), k=-1).astype(bool))
    .stack()
    .reset_index()
)
corr_pairs.columns = ['Feature 1', 'Feature 2', 'Correlation']
corr_pairs = corr_pairs.sort_values('Correlation', key=abs, ascending=False)

print('All pairwise correlations:')
print(corr_pairs.to_string(index=False))

high_pos = corr_pairs[corr_pairs['Correlation'] > 0.5]
high_neg = corr_pairs[corr_pairs['Correlation'] < -0.3]

print('\nHighly Positive Correlations (> 0.5):')
print(high_pos if not high_pos.empty else 'None found')

print('\nNegatively Correlated Features (< -0.3):')
print(high_neg if not high_neg.empty else 'None found')
```

## 9. Feature Engineering

### 9.1 cost_per_person

```python
df['cost_per_person'] = df['cost_for_two'] / 2
print('cost_per_person created.')
print(df['cost_per_person'].describe().round(2))
```

### 9.2 rating_category

```python
def rate_category(r):
    if r < 3.0:   return 'Poor'
    elif r < 3.5: return 'Average'
    elif r < 4.0: return 'Good'
    else:         return 'Excellent'

df['rating_category'] = df['rate'].apply(rate_category)
print('rating_category distribution:')
print(df['rating_category'].value_counts())

fig, ax = plt.subplots(figsize=(6, 4))
order = ['Poor', 'Average', 'Good', 'Excellent']
sns.countplot(x='rating_category', data=df, order=order,
              palette=['#e15759', '#f28e2b', '#4e79a7', '#59a14f'], ax=ax)
ax.set_title('Restaurants by Rating Category')
ax.set_xlabel('Rating Category')
ax.set_ylabel('Count')
for p in ax.patches:
    ax.annotate(f'{int(p.get_height()):,}',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=9)
plt.tight_layout()
plt.show()
```

### 9.3 popularity_score (composite feature)

```python
# Normalized votes × rating as a composite popularity signal
df['popularity_score'] = (
    (df['votes'] / df['votes'].max()) * df['rate']
).round(4)

print('Top 10 most popular restaurants:')
print(
    df[['name', 'location', 'rate', 'votes', 'popularity_score']]
    .sort_values('popularity_score', ascending=False)
    .head(10)
    .to_string(index=False)
)
```

## 10. Business Insights

Based on the EDA, here are **10 actionable insights**:

1. **Online ordering drives volume** — more than 60% of restaurants accept online orders; those restaurants tend to have slightly higher average ratings, suggesting customers rate the convenience positively.

2. **Table booking correlates with quality** — restaurants that allow table booking consistently achieve higher median ratings, indicating that sit-down dining experiences meet customer expectations better.

3. **BTM is the most competitive area** — with 5,124 listings, BTM Layout has the highest restaurant density. New entrants should differentiate strongly or consider less saturated localities.

4. **Quick Bites dominate supply** — Quick Bites (~19K listings) is the most common restaurant type, creating intense competition in the budget segment. Premium casual dining (~10K listings) may offer better margins.

5. **North Indian and Chinese cuisines are oversaturated** — the top two cuisine combinations are North Indian and North Indian+Chinese. Restaurants offering South East Asian, Mediterranean, or niche cuisines face less direct competition.

6. **Cost sweet spot is ₹200–₹600 for two** — the majority of restaurants cluster in this range, aligning with mass-market demand. Premium restaurants (₹1,000+) are rare, signalling an underserved upscale segment.

7. **Votes and rating are positively correlated** — highly voted restaurants tend to maintain higher ratings, confirming that popularity reinforces perceived quality. New restaurants should aggressively solicit early reviews.

8. **Most restaurants are rated 'Good' (3.5–4.0)** — the distribution is left-skewed with few 'Excellent' (>4.0) restaurants. Achieving an excellent rating is a strong differentiator in a crowded market.

9. **High cost ≠ high rating** — the scatter plot shows weak positive correlation between cost and rating. Some affordable restaurants outperform expensive ones, so quality execution matters more than pricing strategy.

10. **Table booking restaurants charge more** — restaurants with table booking have higher average costs, suggesting Zomato could promote a 'premium dining' filter to help customers find upscale experiences easily.
